# 各地域の暖冷房負荷

```python
def draw_integrated_graph(labels, cases):
    # Case is the list of integer, the length of which should be 12.
    
    # number of bars
    n = len(labels)
    
    # heating load, MJ/year
    ys_h = []
    # sensible cooling load, MJ/year
    ys_cs = []
    # latent cooling load, MJ/year
    ys_cl = []
    
    for case in cases:
        l_h, l_cs, l_cl = cs.get_load(region=case[0], insulation=case[1], solar_gain=case[2],
                                      a_mr=case[3], a_or=case[4], a_a=case[5], r_env=case[6])
        ys_h.append(np.sum(l_h))
        ys_cs.append(np.sum(l_cs))
        ys_cl.append(np.sum(l_cl))
    
    fig = plt.figure(figsize=(15,4))
    ax = fig.add_subplot(1,1,1)

    ax.bar(np.arange(n), ys_h, width=0.3, color='red', label='heating')
    ax.bar(np.arange(n)+0.3, ys_cs, width=0.3, color='blue', label='sensible cooling')
    ax.bar(np.arange(n)+0.3, ys_cl, bottom=ys_cs, width=0.3, color='green', label='latent cooling')

    ax.set_xticks(np.arange(n))
    ax.set_xticklabels(labels, rotation=30)
    ax.set_ylabel('load, MJ/year')
    plt.legend()
    plt.show()
```
